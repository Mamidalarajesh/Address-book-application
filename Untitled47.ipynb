```python
import os

def display_menu():
    print("\nAddress Book Menu:")
    print("1. Add Contact")
    print("2. View Contacts")
    print("3. Update Contact")
    print("4. Delete Contact")
    print("5. Exit")

def add_contact():
    name = input("Enter the name: ")
    phone = input("Enter the phone number: ")
    email = input("Enter the email address: ")

    contact = f"{name},{phone},{email}\n"

    with open("address_book.txt", "a") as file:
        file.write(contact)

    print(f"Contact for {name} has been added.")

def view_contacts():
    with open("address_book.txt", "r") as file:
        contacts = file.readlines()

    if not contacts:
        print("No contacts found.")
    else:
        print("\nList of Contacts:")
        for contact in contacts:
            name, phone, email = contact.strip().split(',')
            print(f"Name: {name}")
            print(f"Phone: {phone}")
            print(f"Email: {email}")
            print()

def update_contact():
    name_to_update = input("Enter the name of the contact to update: ")

    with open("address_book.txt", "r") as file:
        contacts = file.readlines()

    updated_contacts = []
    found = False

    for contact in contacts:
        name, phone, email = contact.strip().split(',')
        if name == name_to_update:
            new_phone = input("Enter the new phone number: ")
            new_email = input("Enter the new email address: ")
            contact = f"{name},{new_phone},{new_email}\n"
            found = True
            print(f"Contact for {name} has been updated.")
        updated_contacts.append(contact)

    if not found:
        print(f"No contact found with the name {name_to_update}.")

    with open("address_book.txt", "w") as file:
        file.writelines(updated_contacts)

def delete_contact():
    name_to_delete = input("Enter the name of the contact to delete: ")

    with open("address_book.txt", "r") as file:
        contacts = file.readlines()

    updated_contacts = []
    found = False

    for contact in contacts:
        name, _, _ = contact.strip().split(',')
        if name == name_to_delete:
            found = True
            print(f"Contact for {name} has been deleted.")
        else:
            updated_contacts.append(contact)

    if not found:
        print(f"No contact found with the name {name_to_delete}.")

    with open("address_book.txt", "w") as file:
        file.writelines(updated_contacts)

while True:
    display_menu()
    choice = input("Enter your choice: ")

    if choice == '1':
        add_contact()
    elif choice == '2':
        view_contacts()
    elif choice == '3':
        update_contact()
    elif choice == '4':
        delete_contact()
    elif choice == '5':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")

```
